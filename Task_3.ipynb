## In the following cell we are going to import three libraries named Matplotlib.pyplot, Pandas and Seaborn

```python
import matplotlib.pyplot as plt
# In the above code we have imported matplotlib.pyplot and changed its name to plt
import pandas as pd
# In the above code we have imported pandas and changed its name to pd
import seaborn as sns
# In the above code we have imported seaborn and changed its name to sns
```

```python
tips_df = pd.read_csv("tips.csv")
# In the above code we have called pandas and have created a dataframe of "tips.csv" and stored it in tips_df
```

# Types of  Plots in Seaborn

## 1- lmplot

```python
sns.lmplot(data=tips_df, x="total_bill", y="tip",hue='sex', height=4)
```

### In the above code we are calling seaborn and plotting a lmplot of tips_df with total_bill on x-axis and tip on y-axis with hieght of the graph as 4 and colours of the plotted dots changing according to sex

## 2- Barplot

```python
sns.barplot(x="sex", y="tip",hue="day", data=tips_df)
```

### In the above code we are calling seaborn and plotting a barplot of tips_df with sex on x-axis and tip on y-axis and colours of the plotted bars changing according to day

## 3- Scatterplot

```python
sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips_df)
```

### In the above code we are calling seaborn and plotting a scatterplot of tips_df with total_bill on x-axis and tip on y-axis with colours of the plotted dots changing according to day 

## 4- Lineplot

```python
sns.lineplot(x="total_bill", y="tip", hue="sex", data=tips_df)
```

### In the above code we are calling seaborn and plotting a lineplot of tips_df with total_bill on x-axis and tip on y-axis with colours of the plotted line changing according to sex

## 5- Catplot

```python
sns.catplot(x="day", y="total_bill", hue="sex", data=tips_df, kind="bar")
```

### In the above code we are calling seaborn and plotting a catplot of tips_df with day on x-axis and total_bill on y-axis with colours of the plotted bar changing according to sex

## 6- Boxplot

```python
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips_df)
```

### In the above code we are calling seaborn and plotting a boxplot of tips_df with day on x-axis and total_bill on y-axis with colours of the plotted boxs changing according to sex 

## 7- Violinplot 

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips_df)
```

### In the above code we are calling seaborn and plotting a violinplot of tips_df with day on x-axis and total_bill on y-axis with colours of the plot changing according to sex

## 8- Jointplot 

```python
sns.jointplot(x="total_bill", y="tip", data=tips_df)
```

### In the above code we are calling seaborn and plotting a jointplot of tips_df with total_bill on x-axis and tip on y-axis

## 9- Pairplot 

```python
sns.pairplot(tips_df, hue="sex")
```

### In the above code we are calling seaborn and plotting a pairplot of tips_df with colours of the plotted points changing according to sex 

## 10- Countplot 

```python
sns.countplot(x="day", data=tips_df)
```

### In the above code we are calling seaborn and plotting a countplot of tips_df with day on x-axis and count on y-axis
